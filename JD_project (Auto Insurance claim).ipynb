### Using inter-quartile range to take out the outliners

```python

for x in ['Age']:
    q75,q25 = np.percentile(df.loc[:,x],[75,25])
    intr_qr = q75-q25
 

    max = q75+(1.5*intr_qr)
    min = q25-(1.5*intr_qr)
 
    df.loc[df[x] < min,x] = np.nan
    df.loc[df[x] > max,x] = np.nan
    
    
df['Age'].isna().sum()
```

```python
age_mean = int(df['Age'].loc[df['Age'].isnull() == False].mean())
age_mean
```

```python
df.Age = df.Age.fillna(age_mean)
df.Age = df.Age.astype(int)
df.Age.unique()
```

```python
gmap = {'Male':'Male', 'Female':'Female','Entity': 'Others', 'Joint Gender':'Others', 'NOT STATED':'Others','NO GENDER':'Others', 'SEX':'Others'} 

df['Gender'] = df['Gender'].map(gmap)

df['Gender'].unique()
```

```python
# balancing dataset target


from sklearn.utils import resample

df_maj = df[df['target'] == 0] 
df_min = df[df['target'] == 1]

# upsampling the miniority class
df_min_upsampled= resample(df_min,
                            replace=True,
                            n_samples=10623,
                            random_state=123)

df=pd.concat([df_maj ,df_min_upsampled])
```

## Applying one hot encoding using the pandas library

> NOTE: you can use one hot encoding library from sklearn.preprocessing but i used the pandas libraries (pd.get_dummies)

```python
for col in cat_col:
    df = pd.concat([df, pd.get_dummies(df[col], prefix=col[:3], drop_first=True)], axis=1)
    df.drop(col,axis=1,inplace=True)
    
df
```

## Machine Learning



```python
X = df.drop(['target','ID'], axis=1)
y = df['target']

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=123)
```

```python
display(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier

# 2. Create an instance of the estimator
KNN = KNeighborsClassifier()

# 3. Use the training data to train the estimator
KNN.fit(X_train, y_train)

# 4. Evaluate the model
y_pred= KNN.predict(X_test)

#Confusion matrix

print(confusion_matrix(y_pred,y_test))
print("Classification Report")
print(classification_report(y_test, y_pred))
```

```python
# 1. Import the estimator object (model)
from sklearn.tree import DecisionTreeClassifier
# 2. Create an instance of the estimator
DTC = DecisionTreeClassifier(random_state=100)

# 3. Use the trainning data to train the estimator
DTC.fit(X_train, y_train)

# 4. Evaluate the model
y_pred= DTC.predict(X_test)

#Confusion matrix

print(confusion_matrix(y_pred,y_test))
print("Classification Report")
print(classification_report(y_test, y_pred))
```
